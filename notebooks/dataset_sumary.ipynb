# Dataset Summary

This notebook performs an exploratory data analysis (EDA) on the cleaned retail dataset to understand purchasing behavior and patterns.

- **Insights generated**:
  - Number of customers, products, and transactions
  - Distribution of basket sizes (items per invoice)
  - Most frequently purchased items
  - Temporal trends in purchases
- **Purpose**:
  - Build understanding before rule mining and model training

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/transactions_fpgrowth.csv")
```

```python
lengths = df["items"].str.split(",").apply(len)
```

```python
plt.figure(figsize=(8,4))
sns.histplot(lengths, bins=20, kde=True)
plt.title("Phân phối số mặt hàng mỗi hóa đơn")
plt.xlabel("Số lượng item")
plt.ylabel("Số giao dịch")
plt.show()
```

## Conclusion

This high‑level EDA outlines customer counts, basket sizes, top‑selling items, and purchase timing, giving us a clear picture of shopping behavior. These insights inform feature engineering and parameter choices for both rule‑based mining and deep‑learning recommender models.
